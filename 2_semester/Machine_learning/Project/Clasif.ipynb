```python
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
```

# Clasificación binaria

```python
df=pd.read_csv('bank-full.csv',sep=';')
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.drop(columns=['day','month'])
```

```python
f, ax = plt.subplots(figsize=(14,9)) #subplots and zise
sns.countplot(x="y", data=df)
```

```python
df2=pd.get_dummies(df)
```

```python
df2.info()
```

```python
df2.head()
```

```python
# datos para el modelo
y=df2['y_yes'].values
X=df2.drop(columns=['y_yes','y_no','job_unknown','marital_divorced','education_unknown','default_no','housing_no','loan_no','contact_unknown','poutcome_unknown']).values
```

```python
from datetime import datetime
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# Con XGBoost

https://xgboost.readthedocs.io/en/latest/python/python_api.html#module-xgboost.sklearn

```python
startTime = datetime.now()

gbm_param_grid={'learning_rate':[0.1],
               'n_estimators':[300],
               'subsample':[0.9],
               'gamma':[0,100,200,300],
               'lambda':[100,200,300,400]}


gbm=xgb.XGBClassifier(seed=123)

grid_mse=GridSearchCV(estimator=gbm,param_grid=gbm_param_grid,scoring='f1',cv=5,return_train_score=True,n_jobs=-1)

grid_mse.fit(X_train,y_train)





print(datetime.now() - startTime)
```

```python
# Print the tuned parameters and score
print("Tuned xgb Parameters: {}".format(grid_mse.best_params_)) 
print("Best score is {}".format(grid_mse.best_score_))
```

```python
# predecir datos de prueba
y_pred_train_xgb = grid_mse.predict(X_train)
y_pred_test_xgb = grid_mse.predict(X_test)

y_pred_prob_xgb = grid_mse.predict_proba(X_test)[:,1] # Compute predicted probabilities: y_pred_prob
```

```python
print('Train')
print(classification_report(y_train,y_pred_train_xgb))
print()
print('Test')
print(classification_report(y_test,y_pred_test_xgb))
```

```python
ax = plt.axes()
sns.heatmap(confusion_matrix(y_train,y_pred_train_xgb), annot=True,  ax=ax, cmap="coolwarm",fmt='.2f', linewidths=.05)
ax.set_title('Confusion matrix Train Heatmap (XGB)')
```

```python
ax = plt.axes()
sns.heatmap(confusion_matrix(y_test,y_pred_test_xgb), annot=True,  ax=ax, cmap="coolwarm",fmt='.2f', linewidths=.05)
ax.set_title('Confusion matrix Test Heatmap (XGB)')
```

```python
AUCSxgb=round(roc_auc_score(y_test, y_pred_prob_xgb),3) # Compute  AUC score
fprxgb, tprxgb, thresholdsxgb = roc_curve(y_test , y_pred_prob_xgb) # Generate ROC curve values: fpr, tpr, thresholds
```

```python
scores_mean = grid_mse.cv_results_['mean_test_score']
scores_sd = grid_mse.cv_results_['std_test_score']
scores_mean_tr = grid_mse.cv_results_['mean_train_score']
scores_sd_tr = grid_mse.cv_results_['std_train_score']
```

```python
plt.style.use('seaborn-whitegrid')
x_ax = list(range(0,len(scores_mean)))
plt.rcParams['figure.figsize'] = (12, 6)
plt.fill_between(x_ax, scores_mean - scores_sd, scores_mean + scores_sd, alpha=0.1, color='blue')
plt.plot(x_ax, scores_mean, color='black',label='test')
plt.scatter(np.argmax(scores_mean), scores_mean[np.argmax(scores_mean),], s=300, linewidth=2, facecolors = 'none', edgecolors='red')
plt.errorbar(x_ax, scores_mean_tr, yerr=scores_sd_tr, fmt='-o', color='gray', ecolor='gray',label='train')
plt.xlabel('Modelo',size=15)
plt.ylabel('Score: f1',size=15)
plt.suptitle('Grid Search CV',size=20)
txt = 'Mejor modelo, XGB con learning_rate ' + str(grid_mse.best_estimator_.learning_rate) + ' ,n_estimators ' + str(grid_mse.best_estimator_.n_estimators)+ ' ,gamma ' + str(grid_mse.best_estimator_.gamma)+' ,lambda 400'
plt.title(txt,size=15);
plt.legend(loc="best");
plt.ylim([0.0,1+0.05])
```

# Características más importantes

```python
# Create the DMatrix: housing_dmatrix
dmatrix = xgb.DMatrix(data=df2.drop(columns=['y_yes','y_no','job_unknown','marital_divorced','education_unknown','default_no','housing_no','loan_no','contact_unknown','poutcome_unknown']), label=df2['y_yes'])

# Create the parameter dictionary: params
params = {"objective":"binary:logistic", 'gamma': 0, 'lambda': 400, 'learning_rate': 0.1, 'n_estimators': 300, 'subsample': 0.9}

# Train the model: xg_reg
xg_reg = xgb.train(params=params, dtrain=dmatrix, num_boost_round=10)

# Plot the feature importances
xgb.plot_importance(xg_reg,max_num_features=10,importance_type='gain', xlabel='Gain')
plt.show()
```

# Con RegLog

```python
startTime = datetime.now()

# Setup the hyperparameter grid
#c_space = np.logspace(-5, 8, 16)
c_space =[0.001,0.01,0.1,1,2,4,6,8,10,100,1000,1e4,1e5,1e6,1e7,1e8]
param_grid = {'C': c_space}

logreg = LogisticRegression(solver='sag',max_iter=500) # Create the classifier: logreg

# Instantiate the GridSearchCV object: logreg_cv
logreg_cv = GridSearchCV(logreg, param_grid, cv=5,scoring='f1',return_train_score=True,n_jobs=-1)

logreg_cv.fit(X_train,y_train)

print(datetime.now() - startTime)
```

```python
# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_)) 
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
y_pred_train_log = logreg_cv.predict(X_train) # Predict the labels of the test set: y_pred
y_pred_test_log = logreg_cv.predict(X_test)

y_pred_prob_log = logreg_cv.predict_proba(X_test)[:,1] # Compute predicted probabilities: y_pred_prob
```

```python
print('Train')
print(classification_report(y_train,y_pred_train_log))
print()
print('Test')
print(classification_report(y_test,y_pred_test_log))
```

```python
ax = plt.axes()
sns.heatmap(confusion_matrix(y_train,y_pred_train_log), annot=True,  ax=ax, cmap="coolwarm",fmt='.2f', linewidths=.05)
ax.set_title('Confusion matrix Train Heatmap (LogReg)')
```

```python
ax = plt.axes()
sns.heatmap(confusion_matrix(y_test,y_pred_test_log), annot=True,  ax=ax, cmap="coolwarm",fmt='.2f', linewidths=.05)
ax.set_title('Confusion matrix Test Heatmap (LogReg)')
```

```python
AUCSLR=round(roc_auc_score(y_test, y_pred_prob_log),3) # Compute  AUC score
fprlog, tprlog, thresholdslog = roc_curve(y_test , y_pred_prob_log) # Generate ROC curve values: fpr, tpr, thresholds
```

```python
# # Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fprxgb, tprxgb, label='XGB, AUCS:'+str(AUCSxgb))
plt.plot(fprgb, tprgb, label='GB, AUCS:'+str(AUCSgb))
plt.plot(fprlog, tprlog, label='LogReg, AUCS:'+str(AUCSLR))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve ')
plt.legend(loc='upper right')
plt.show()
```
