```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import median_absolute_error
```

# 1) Replicar resultados articulo

```python
train=pd.read_csv('train.csv')
```

```python
train.head()
```

```python
# datos para el modelo
y=train.critical_temp.values
X=train.drop(columns=['critical_temp']).values
```

```python
from datetime import datetime
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# Con XGBoost

https://xgboost.readthedocs.io/en/latest/python/python_api.html#module-xgboost.sklearn

```python
startTime = datetime.now()

gbm_param_grid={'learning_rate':[0.01,0.1,0.5,0.9],
               'n_estimators':[50,100,200,300],
               'subsample':[0.9]}


gbm=xgb.XGBRegressor()

grid_mse=GridSearchCV(estimator=gbm,param_grid=gbm_param_grid,scoring='neg_mean_squared_error',cv=5,return_train_score=True,n_jobs=-1)

grid_mse.fit(X_train,y_train)





print(datetime.now() - startTime)
```

```python
# Print the tuned parameters and score
print("Tuned xgb Parameters: {}".format(grid_mse.best_params_)) 
print("Best score is {}".format(grid_mse.best_score_))
```

```python
# predecir datos de prueba
y_pred_train_xgb = grid_mse.predict(X_train)
y_pred_test_xgb = grid_mse.predict(X_test)
y_pred = grid_mse.predict(X)
```

```python
mse=mean_squared_error(y_test,y_pred_test_xgb)
r2=r2_score(y_test,y_pred_test_xgb)
mae=median_absolute_error(y_test,y_pred_test_xgb)
```

```python
print((mse)**(0.5),r2,mae)
```

```python
# Figura 8
xl=[0,200]
yl=[0,200]
#plt.figure(figsize = (15,10))
plt.plot(xl,yl, color='red')
plt.scatter(y,y_pred,color='black')
plt.xlim([0,190])
plt.ylim([0,155])
plt.xlabel('Observed Critial Temperature (K)'), plt.ylabel('Predicted Critial Temperature (K)'), plt.title('XGBoost ')
plt.show()
```

```python
scores_mean = grid_mse.cv_results_['mean_test_score']
scores_sd = grid_mse.cv_results_['std_test_score']
scores_mean_tr = grid_mse.cv_results_['mean_train_score']
scores_sd_tr = grid_mse.cv_results_['std_train_score']
```

```python
plt.style.use('seaborn-whitegrid')
x_ax = list(range(0,len(scores_mean)))
plt.rcParams['figure.figsize'] = (12, 6)
plt.fill_between(x_ax, scores_mean - scores_sd, scores_mean + scores_sd, alpha=0.1, color='blue')
plt.plot(x_ax, scores_mean, color='black',label='test')
plt.scatter(np.argmax(scores_mean), scores_mean[np.argmax(scores_mean),], s=300, linewidth=2, facecolors = 'none', edgecolors='red')
plt.errorbar(x_ax, scores_mean_tr, yerr=scores_sd_tr, fmt='-o', color='gray', ecolor='gray',label='train')
plt.xlabel('Modelo',size=15)
plt.ylabel('Score: MSE',size=15)
plt.suptitle('Grid Search CV',size=20)
txt = 'Mejor modelo, XGB con learning_rate ' + str(grid_mse.best_estimator_.learning_rate) + ' y n_estimators ' + str(grid_mse.best_estimator_.n_estimators)
plt.title(txt,size=15);
plt.legend(loc="best");
#plt.ylim([0.3,1+0.05])
```

# Características más importantes

```python
# Create the DMatrix: housing_dmatrix
dmatrix = xgb.DMatrix(data=train.drop(columns=['critical_temp']), label=train.critical_temp)

# Create the parameter dictionary: params
params = {"objective":"reg:squarederror", "n_estimators":300,'learning_rate': 0.1,'subsample': 0.9}

# Train the model: xg_reg
xg_reg = xgb.train(params=params, dtrain=dmatrix, num_boost_round=10)

# Plot the feature importances
xgb.plot_importance(xg_reg,max_num_features=20,importance_type='gain', xlabel='Gain')
plt.show()
```

# Efecto de gamma y lambda en el arbol

```python
from matplotlib.pylab import rcParams
from ipywidgets import interact
def EjemParam(max_depth,gamma,lambdap):
    # Create the DMatrix: housing_dmatrix
    dmatrix = xgb.DMatrix(data=train[['range_ThermalConductivity','wtd_gmean_ThermalConductivity']], label=train.critical_temp)

    # Create the parameter dictionary: params
    params = {"objective":"reg:squarederror", "n_estimators":300,'learning_rate': 0.1,'subsample': 0.9,'max_depth':max_depth, 'gamma':gamma,'lambda':lambdap}

    # Train the model: xg_reg
    xg_reg = xgb.train(params=params, dtrain=dmatrix, num_boost_round=10)

    ##set up the parameters
    rcParams['figure.figsize'] = 40,50
    # Plot the fifth tree

    xgb.plot_tree(xg_reg, num_trees=1, rankdir='LR')
    plt.show()
    return()
```

```python
interact(EjemParam,max_depth=[5],gamma=[0,50],lambdap=[0,50])
```

# Con Scikit-learn

```python
# Import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingRegressor

startTime = datetime.now()

gbm_param_grid={'learning_rate':[0.01,0.1,0.5,0.9],'n_estimators':[50,100,200,300]}


# Instantiate gb
gb = GradientBoostingRegressor()

gb_cv=GridSearchCV(estimator=gb,param_grid=gbm_param_grid,scoring='neg_mean_squared_error',cv=5,return_train_score=True,n_jobs=-1)

gb_cv.fit(X_train,y_train)





print(datetime.now() - startTime)
```

```python
# Print the tuned parameters and score
print("Tuned gb Parameters: {}".format(gb_cv.best_params_)) 
print("Best score is {}".format(gb_cv.best_score_))
```

```python
# predecir datos de prueba
y_pred_train_gb = gb_cv.predict(X_train)
y_pred_test_gb = gb_cv.predict(X_test)
y_pred = gb_cv.predict(X)
```

```python
mse=mean_squared_error(y_test,y_pred_test_gb)
r2=r2_score(y_test,y_pred_test_gb)
mae=median_absolute_error(y_test,y_pred_test_gb)
```

```python
print((mse)**(0.5),r2,mae)
```

```python
# Figura 8
xl=[0,200]
yl=[0,200]
#plt.figure(figsize = (15,10))
plt.plot(xl,yl, color='red')
plt.scatter(y,y_pred,color='black')
#plt.xlim([0,190])
#plt.ylim([-30,155])
plt.xlabel('Observed Critial Temperature (K)'), plt.ylabel('Predicted Critial Temperature (K)'), plt.title('GradientBoost ')
plt.show()
```
